# This is a sample code in jupyter notebook to clean data.

Import library

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

Load Data

```python
hazardata = pd.read_csv("hazard-event-data.csv", thousands=',')
```

```python
hazardata.head()
```

Summarize damage by region 

```python
hazardata["Property Damage"].groupby(hazardata["Region"]).mean()
```

Summarize by year and event type

```python
hyear=hazardata.groupby('Year',as_index=False)[['Fatalities','Injuries','Property Damage']].sum()
print(hyear.reset_index())
```

```python
hazardata.groupby('Hazard Event')[['Fatalities','Injuries','Property Damage']].sum()
```

You may also import matplotlib to make plots (pip install). 
For example, plot the property damage and fatalities/injuries across years

```python
import matplotlib.pyplot as plt
plt.plot(hyear['Year'],hyear['Property Damage'])
plt.show()
```

```python
plt.scatter(hyear['Year'],hyear['Injuries']+hyear['Fatalities'])
plt.show()
```
